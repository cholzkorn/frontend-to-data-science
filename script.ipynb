# Data Import

```r
df <- read.csv("users.csv", sep=";")
```

```r
summary(df)
```

```r
ggplot(df, aes(x = salary, y = satisfaction)) + 
  geom_point()
```

# Linear Model

```r
linear <- lm(satisfaction ~ salary, data=df)
```

```r
summary(linear)
```

```r
library(ggplot2)

ggplot(df, aes(x = salary, y = satisfaction)) + 
  geom_point() +
  stat_smooth(method = "lm", col = "red")
```

# Export Plot

```r
ggsave("linear_regression.png", plot = last_plot())
```
